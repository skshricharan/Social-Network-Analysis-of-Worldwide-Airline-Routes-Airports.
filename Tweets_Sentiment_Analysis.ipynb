```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df =pd.read_csv('D:\Documents\VIT\SIN\Project/Tweets.csv')
df.head()
```

```python
# Unique values of sentiment
df['airline_sentiment'].unique()
```

```python
# Unique values of sentiment plot

ax = sns.countplot(x="airline_sentiment", data=df)
```

```python
# Unique values of airline

plt.figure(figsize=(10,10))
ax = sns.countplot(x="airline", data=df)
```

```python
df['tweet_length'] = df['text'].apply(len)
df.groupby(['tweet_length','airline_sentiment']).size().unstack().plot(kind='line',stacked=False)
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(12,10)})
sns.countplot(x="airline",hue="airline_sentiment",data=df)
plt.title("Airline sentiment for each US Airline")
```

```python
df[['tweet_length','airline_sentiment','airline_sentiment_confidence']].groupby(['tweet_length','airline_sentiment']).mean().unstack().plot(kind='line',stacked=False)
plt.title('Average Airline Sentiment vs Tweet Length')
```

```python
df[['tweet_length','airline_sentiment','airline_sentiment_confidence']].groupby(['tweet_length','airline_sentiment']).median().unstack().plot(kind='line',stacked=False)
plt.title('Median Airline Sentiment vs Tweet Length')
```
